# Comment ajouter des colonnes

```python
import pandas as pd
df = pd.read_csv("data.csv")
df
```

```python
# on va ajouter ici une colonne calculant le prix final
# on va remplacer les valeurs manquantes par 0 pour pouvoir faire le calcul
```

```python
df["tax"].fillna(0, inplace=True)
```

```python
df
```

```python
# pour voir s'il ya des valeurs nulles dans la colonne price_paid
#df["price_paid"].isnull()
```

```python
df.price_paid = df.price_paid.apply(lambda x: x.replace("$", "")) # pour retirer les $ dans la colonne price_paid
df.price_paid = df.price_paid.astype(float) # pour convertir le type de la variable price_paid en float
```
